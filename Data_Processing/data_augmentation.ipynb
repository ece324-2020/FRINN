Code for Data Augmentation.
Techniques used are:


*   Horizontal flip
*   Vertical flip
*   Rotation by degrees (15, 30, 45, 60, 75)
*   Gaussian blur
*   Salt and Pepper

Each transform is defined and performed on a dataset object. The augmented images are stored in a zip file and downloaded.

```
!pip install git+https://github.com/aleju/imgaug
from imgaug import augmenters as iaa
import imgaug as ia
import numpy as np
from PIL import Image
```

```
class ImgAugTransform_hflip:
  def __init__(self):
    self.aug = iaa.Sequential([
        iaa.Fliplr(1.0)
    ])
      
  def __call__(self, img):
    img = np.array(img)
    return self.aug.augment_image(img)

transforms_hflip = ImgAugTransform_hflip()
dataset_hflip = torchvision.datasets.ImageFolder(root='/content/gdrive/My Drive/school/augmentation_test', transform=transforms_hflip)
```

```
%cd ur_s-hflip
img_num = 0
for img, label in dataset_hflip:
  im = Image.fromarray(img)
  im.save('ur_s-hflip-'+str(img_num)+'.png')
  img_num+=1 

%cd - 
!zip -r ur_s-hflip.zip ur_s-hflip
#Download files
files.download('ur_s-hflip.zip')
```
